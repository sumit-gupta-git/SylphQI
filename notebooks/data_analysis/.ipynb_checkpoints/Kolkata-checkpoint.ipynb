```python
#importing dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
#reading data from city_day.csv taken from kaggle from 01-01-2015 to 31-12-2024 of various Indian Cities
df = pd.read_csv("India_Cities_AQI_Weather_2015_2024_Combined.csv")
```

```python
#converting date column into datetime object
df["Date"]=pd.to_datetime(df["Date"])
```

```python
#diving group based on city names
group_df = df.groupby("City")

#making city wise dataframe
kolkata_df = group_df.get_group("Kolkata")
```

```python
kolkata_df
```

```python
#reseting indices
kolkata_df = kolkata_df.reset_index(drop=True)
```

```python
kolkata_df
```

### Forecasting using FB Prophet

```python
kolkata_df
```

```python
kolkata_df.columns
```

```python
kolkata_df.index
```

```python
#splitting each pollutant
#pm2.5
kolkata_pm25 = pd.DataFrame(kolkata_df[["Date", "PM2.5"]])
kolkata_pm25.columns = ["ds", "y"]

#pm10
kolkata_pm10 = pd.DataFrame(kolkata_df[["Date", "PM10"]])
kolkata_pm10.columns = ["ds", "y"]

#NO2
kolkata_no2 = pd.DataFrame(kolkata_df[["Date", "NO2"]])
kolkata_no2.columns = ["ds", "y"]

#NH3
kolkata_nh3 = pd.DataFrame(kolkata_df[["Date", "NH3"]])
kolkata_nh3.columns = ["ds", "y"]

#CO
kolkata_co = pd.DataFrame(kolkata_df[["Date", "CO"]])
kolkata_co.columns = ["ds", "y"]

#SO2
kolkata_so2 = pd.DataFrame(kolkata_df[["Date", "SO2"]])
kolkata_so2.columns = ["ds", "y"]

#O3
kolkata_o3 = pd.DataFrame(kolkata_df[["Date", "O3"]])
kolkata_o3.columns = ["ds", "y"]
```

```python
kolkata_pm25['ds']
```

```python
from prophet import Prophet
```

```python
#initialize prophet for each pollutant
model_pm25 = Prophet()
model_pm25.fit(kolkata_pm25)

#future timestamps
future_pm25 = model_pm25.make_future_dataframe(periods=365)

#predictions
forecast_pm25 = model_pm25.predict(future_pm25)

#print results
print(forecast_pm25[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
model_pm10 = Prophet()
model_pm10.fit(kolkata_pm10)

#future timestamps
future_pm10 = model_pm10.make_future_dataframe(periods=365)

#predictions
forecast_pm10 = model_pm10.predict(future_pm10)

#print results
print(forecast_pm10[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```
