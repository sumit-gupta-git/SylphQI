```python
#importing dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
#reading data from city_day.csv taken from kaggle from 01-01-2015 to 31-12-2024 of various Indian Cities
df = pd.read_csv("India_Cities_AQI_Weather_2015_2024_Combined.csv")
```

```python
df.head()
```

```python
#converting date column into datetime object
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
#diving group based on city names
group_df = df.groupby("City")

#making city wise dataframe
mumbai_df = group_df.get_group("Mumbai")
```

```python
mumbai_df.head()
```

```python
#reseting indices
mumbai_df = mumbai_df.reset_index(drop=True)
```

```python
mumbai_df.head()
```

## Forecasting using FB Prophet

```python
#splitting each pollutant
#pm2.5
mumbai_pm25 = pd.DataFrame(mumbai_df[["Date", "PM2.5"]])
mumbai_pm25.columns = ["ds", "y"]

#pm10
mumbai_pm10 = pd.DataFrame(mumbai_df[["Date", "PM10"]])
mumbai_pm10.columns = ["ds", "y"]

#NO2
mumbai_no2 = pd.DataFrame(mumbai_df[["Date", "NO2"]])
mumbai_no2.columns = ["ds", "y"]

#NH3
mumbai_nh3 = pd.DataFrame(mumbai_df[["Date", "NH3"]])
mumbai_nh3.columns = ["ds", "y"]

#CO
mumbai_co = pd.DataFrame(mumbai_df[["Date", "CO"]])
mumbai_co.columns = ["ds", "y"]

#SO2
mumbai_so2 = pd.DataFrame(mumbai_df[["Date", "SO2"]])
mumbai_so2.columns = ["ds", "y"]

#O3
mumbai_o3 = pd.DataFrame(mumbai_df[["Date", "O3"]])
mumbai_o3.columns = ["ds", "y"]
```

```python
from prophet import Prophet
```

```python
#initialize prophet for each pollutant
model_pm25 = Prophet()
model_pm25.fit(mumbai_pm25)

#future timestamps
future_pm25 = model_pm25.make_future_dataframe(periods=365)

#predictions
forecast_pm25 = model_pm25.predict(future_pm25)

#print results
print(forecast_pm25[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```
